### Análisis del comportamiento de contaminantes
Este cuaderno grafica el comportamiento de los contaminantes, determina la fecha con menor y mayor concentración y muestra esos puntos resaltados.

```python
import pandas as pd
import matplotlib.pyplot as plt

### --- Cargar datos ---
# Cambia la ruta del archivo por la ruta correcta dentro de tu repositorio
df = pd.read_csv('informe_sisaire.csv')  # Ajusta el nombre si tu archivo cambia

### --- Asegurar que la columna de fecha esté en formato datetime ---
df['fecha'] = pd.to_datetime(df['fecha'])

### --- Identificar fecha con menor y mayor concentración ---
col = 'contaminante'  # Cambiar al nombre real de tu columna de contaminante
min_row = df.loc[df[col].idxmin()]
max_row = df.loc[df[col].idxmax()]

print('Fecha con menor concentración:')
print(min_row['fecha'], '--', min_row[col])
print('\nFecha con mayor concentración:')
print(max_row['fecha'], '--', max_row[col])

### --- Graficar comportamiento ---
plt.figure(figsize=(12,6))
plt.plot(df['fecha'], df[col], label='Concentración', linewidth=2)

# Punto mínimo
plt.scatter(min_row['fecha'], min_row[col], s=100, edgecolor='black', label='Mínimo')

# Punto máximo
plt.scatter(max_row['fecha'], max_row[col], s=100, edgecolor='black', label='Máximo')

plt.title('Comportamiento del contaminante')
plt.xlabel('Fecha')
plt.ylabel('Concentración')
plt.legend()
plt.grid(True)
plt.show()
```
